```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
k = np.arange(1, n+1) # number of heads in n independent coin tosses
p = 0.5

probabilites = binom.pmf(k, n, p)
probabilites
```

```python
plt.figure(figsize=(8, 5))
plt.bar(k, probabilites)
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Probability distribution of number of heads in n independent coin tosses', fontsize = 11)
plt.show()
```
